```python
#Cleaning
selected_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']
X = house_df[selected_features]
X
```

```python
y=house_df['price']
y
```

```python
X.shape
```

```python
y.shape
```

```python
#We are doing this so that all variables are of the same scale, and one does not dominate the other
#To train a deep learning model all parameters should be comparable
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#Max value in the data
scaler.data_max_

```

```python
#Min value in the data
scaler.data_min_
```

```python
#reshape the output
y=y.values.reshape(-1,1)
y_scaled=scaler.fit_transform(y)
y_scaled
```

```python
#splitting data in training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#adding layers to the model
model = Sequential()
model.add(Dense(100, input_dim = 7, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(200, activation = 'relu'))
model.add(Dense(1, activation='linear'))
```

```python
#summary of the model
model.summary()
```

```python
#compiling the model
model.compile(optimizer = 'Adam', loss = 'mean_squared_error')
```

```python
# Train the model on the training data
# X_train: The input data for training, containing features like number of bedrooms, square footage, etc.
# y_train: The target data for training, containing the actual house prices.
# epochs=100: The model will go through the entire training dataset 100 times to learn.
# batch_size=50: The model processes 50 samples at a time before updating its parameters.
# validation_split=0.2: 20% of the training data will be used to validate the model's performance after each epoch.

epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, validation_split = 0.2)
```

```python
epochs_hist.history.keys()
```

```python
#Graphing the model's loss during the training
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
#'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement'
#Putting in values of an example house to see what the price would look like
X_test_1 = np.array([[4,3,1960, 5000, 1, 2000, 3000]])

#scaling the data
scaler_1=MinMaxScaler()
X_test_scaled_1= scaler_1.fit_transform(X_test_1)

y_predict_1 = model.predict(X_test_scaled_1)

y_predict_1 = scaler.inverse_transform(y_predict_1)
y_predict_1
```

```python
#Graphing my model predictions from my testing to the true values
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
#As the graph shows the values need improvement, scaling back to the original data
y_predict_orig = scaler.inverse_transform(y_predict)
y_test_orig = scaler.inverse_transform(y_test)
```

```python
#plotting
plt.plot(y_test_orig, y_predict_orig, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.xlim(0,5000000)
plt.ylim(0,3000000)
```

```python
#Generating different metrics
k=X_test.shape[1]
n=len(X_test)
n
```

```python
k
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)), '.3f'))
MSE = mean_squared_error(y_test_orig, y_predict_orig)
MAE = mean_absolute_error(y_test_orig, y_predict_orig)
r2 = r2_score(y_test_orig, y_predict_orig)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE=',MAE, '\nR2=', r2, '\nAdjusted R2=', adj_r2)
```

```python
#Increasing the number of features
selected_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'yr_renovated', 'zipcode', 'lat' , 'long' , 'sqft_living15', 'sqft_lot15', 'view', 'condition', 'grade', 'yr_built', 'waterfront']
X = house_df[selected_features]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
y=house_df['price']
```

```python
#reshape the output
y=y.values.reshape(-1,1)
y_scaled=scaler.fit_transform(y)

#splitting data in training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(10, input_dim = 18, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(200, activation ='relu'))
model.add(Dense(200, activation ='relu'))

model.add(Dense(1, activation='linear'))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
#Training the new model
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, verbose = 1, validation_split = 0.2)
```

```python
#Graphing the model's loss during the training
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
#Graphing my model predictions from my testing to the true values
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel("Model Predictions")
plt.ylabel("True Values (ground Truth)")
plt.title('Linear Regression Predictions')
plt.show()
```

```python
#Scaling back to the original data
y_predict_orig = scaler.inverse_transform(y_predict)
y_test_orig = scaler.inverse_transform(y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)), '.3f'))
MSE = mean_squared_error(y_test_orig, y_predict_orig)
MAE = mean_absolute_error(y_test_orig, y_predict_orig)
r2 = r2_score(y_test_orig, y_predict_orig)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE=',MAE, '\nR2=', r2, '\nAdjusted R2=', adj_r2)
```
